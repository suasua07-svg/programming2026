## 0. 연습용 미니 데이터 (그대로 실행)

```python
import pandas as pd
df = pd.DataFrame({
    "상호명": ["두부집", "윤송헤어", "별난노래방", "세븐일레븐", "김밥나라"],
    "상권업종대분류명": ["음식", "수리·개인", "예술·스포츠", "소매", "음식"],
    "시군구명": ["서구", "유성구", "서구", "중구", "서구"],
})
df
```

## 1. 🐞 버그 ① — 컬럼 이름이 틀렸어요

- **예상 에러**: `KeyError: '업종'`
- **힌트**: 진짜 컬럼 이름은 무엇일까요? (위 데이터를 보세요)

```python
# 🐞 버그: 컬럼 이름이 틀렸습니다
print(df["업종"].value_counts())
```

```python
# ✅ 고친 코드
print(df["상권업종대분류명"].value_counts())
```

## 2. 🐞 버그 ② — 그래프 한글이 깨집니다 (□□□)

- **힌트**: 준비 셀에서 빠뜨린 **한 줄**이 있습니다. (`koreanize...`)

```python
# 🐞 버그: 한글 폰트 설정을 빠뜨렸습니다
import matplotlib.pyplot as plt
df["상권업종대분류명"].value_counts().plot(kind="bar")
plt.title("업종별 가게 수")     # ← 한글이 깨질 수 있음
plt.show()
```

```python
# ✅ 고친 코드
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'DejaVu Sans'
import matplotlib.font_manager as fm
# 한글 폰트 설정 (Mac: AppleGothic, Windows: malgun.ttf, Linux: Noto Sans CJK)
plt.rcParams['axes.unicode_minus'] = False

df["상권업종대분류명"].value_counts().plot(kind="bar", figsize=(8, 5))
plt.title("업종별 가게 수")
plt.xlabel("상권업종대분류명")
plt.ylabel("가게 수")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ✅ 고친 코드
import gradio as gr

def hello(name):
    return f"{name}님 환영합니다"

gr.Interface(fn=hello, inputs="text", outputs="text").launch(share=True)
```
